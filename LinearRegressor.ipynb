# Linear Regressor From Scratch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\PYTHON\FuelConsumption.csv")
data.head(10)
```

```python
x_train = data[["ENGINESIZE"]][:1000].values
```

```python
y_train = data[["CO2EMISSIONS"]][:1000].values
```

```python
class LinearRegressor:
    def __init__(self):
        self.learning_rate = 0.001
        self.iter = 100
        self.intercept = 0
        self.coefficient = 0
        print("Linear Regressor Initialized!")
    

    def fit(self, x, y):
        xbar = np.mean(x)
        ybar = np.mean(y)
        self.coefficient = int(np.sum(((x-xbar) * (y- ybar)))/np.sum(((x-xbar)**2)))
        self.intercept = int(ybar - (xbar * self.coefficient))
 

    def predict(self, x):
        yhat = np.array(self.intercept + (x * self.coefficient), dtype=np.int64)
        return yhat







model = LinearRegressor()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(data[["ENGINESIZE"]][1000:].values)
```

```python
plt.scatter(data[["ENGINESIZE"]][1000:].values, data[["CO2EMISSIONS"]][1000:].values)
plt.plot(data[["ENGINESIZE"]][1000:].values, y_pred)
plt.show()
```
